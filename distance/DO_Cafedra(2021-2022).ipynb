```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
import pytz
import matplotlib as mpl
import calendar
```

```python
dataCafedra = pd.read_csv('C://Users//37529//Desktop//data since//lms//zaochniki//DO_Сafedra(2021-2022).csv', header=0, index_col=0, dayfirst=True)
```

```python
dataCafedra
```

```python
cafedra = [2]
data_102 = dataCafedra[dataCafedra["cafedra"].isin(cafedra)]
data_103=data_102.groupby(['WEEK', 'shortname']).size().reset_index(name='count')
data_104 = pd.pivot_table(data_103, index=['shortname'], columns=['WEEK'], values=['count'])
plt.figure(figsize=(30, 10))
sns.set_theme(style="white")
sns.heatmap(data_104, vmax=3000, vmin=0, cmap="Blues", annot=True, fmt='g', linewidths=.5)
plt.xlabel('week')
plt.ylabel('course');
plt.title('Кафедра электронных вычислительных средств'); 
```
