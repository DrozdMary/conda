```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
import matplotlib.pyplot as plt
```

```python
# Дистанционка (2021-2022). Строю график. 
```

```python
data = pd.read_csv('bd/data_M_H_D.csv', header=0, index_col=0)
```

```python
data
```

```python
month = [6]
data_1 = data[data["month"].isin(month)]
data_2=data_1.groupby(['day', 'hour']).size().reset_index(name='count')
data_3 = pd.pivot_table(data_2, index=['hour'], columns=['day'], values=['count'])
plt.figure(figsize=(20, 10))
sns.set_theme(style="white")
sns.heatmap(data_3, vmax=1000, vmin=0, cmap="Blues", annot=True, fmt='g', linewidths=.5)
plt.xlabel('Day')
plt.ylabel('Hour')
plt.title('Year=2022, Month=6');
```
